```python
import random


words = ['python','program']


a = random.choice(words)


print('The word has', len(a), 'letters')


right = ['__'] * len(a)

wrong = []


def update():
    for i in right:
        print(i, end = ' ')
    print()
    
def parts(x):
    if x==0:
        print('|------------')
        print('|     |      ')
        print('|            ')
        print('|            ')
        print('=============')
    if x==1:
        print('|------------')
        print('|     |      ')
        print('|     O      ')
        print('|            ')
        print('|            ')
        print('|            ')
        print('|============')
    if x==2:
        print('|------------')
        print('|     |      ')
        print('|     O      ')
        print('|     |      ')
        print('|     |      ')
        print('|            ')
        print('|============')
    if x==3:
        print('|------------')
        print('|     |      ')
        print('|     O      ')
        print('|    /|\     ')
        print('|     |      ')
        print('|            ')
        print('|============')
        
    if x==4:
        print('|------------')
        print('|     |      ')
        print('|     O      ')
        print('|    /|\     ')
        print('|     |      ')
        print('|    / \     ')
        print('|            ')
        print('|============')

    
while True:
    print('=============================================================')
    
    

    guess = input('Guess the letter')


    if guess in a:
        index = 0
        for i in a:
            if i == guess:
                right[index] = guess
            index += 1
        update()
        parts(len(wrong))
        
    else:
        if guess not in wrong:
            wrong.append(guess)
            parts(len(wrong))
            print(wrong)
        if len(wrong) > 4:
            print('You Lose')
            break
            
        else:
            print('Already guessed')
    if '__' not in right:
        print('You win')
        break
```
